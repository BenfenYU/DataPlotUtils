```python
import pandas as pd
from utils import unit_convert
import plotly.graph_objects as go


def read_C16_CT3():
    model_ecd_correct_path = r"C:\Users\Jyu20\Downloads\C16_CT3_validate_wellbalance\model_ecd_with_correct_property.csv"
    model_ecd_correct = pd.read_csv(model_ecd_correct_path, skiprows=[1])
    model_ecd_correct = model_ecd_correct.rename(columns={'time':'time_key', 'WellBalance.ATBITECDSIM':'ecd'})

    model_ecd_measure_ecd_path = r"C:\Users\Jyu20\Downloads\C16_CT3_validate_wellbalance\measure_spp_ecd_and_model_with_default_property.csv"
    model_ecd_measure_ecd = pd.read_csv(model_ecd_measure_ecd_path, skiprows=[1])

    measure_ecd = model_ecd_measure_ecd[['UtcTime', 'TELESCOPE.ECD']]
    measure_ecd = measure_ecd.rename(columns={'UtcTime': 'time_key', 'TELESCOPE.ECD': 'ecd'})
    measure_ecd['ecd'] *= unit_convert('g/cm3', 'kg/m3')

    model_ecd_default = model_ecd_measure_ecd[['UtcTime', 'WellBalance.ATBITECDSIM']]
    model_ecd_default = model_ecd_default.rename(columns={'UtcTime': 'time_key', 'WellBalance.ATBITECDSIM': 'ecd'})
    model_ecd_default['ecd'] *= unit_convert('g/cm3', 'kg/m3')


    model_spp_correct_path = r"C:\Users\Jyu20\Downloads\C16_CT3_validate_wellbalance\model_spp_with_correct_property.csv"
    model_spp_correct = pd.read_csv(model_spp_correct_path, skiprows=[1])
    model_spp_correct = model_spp_correct.rename(columns={'time':'time_key', 'WellBalance.SPPSIM':'spp'})

    measure_spp = model_ecd_measure_ecd[['UtcTime', 'StandPipeSystem.PRS']].rename(columns={'UtcTime':'time_key', 'StandPipeSystem.PRS':'spp'})
    measure_spp['spp'] *= 1000

    model_spp_default = model_ecd_measure_ecd[['UtcTime', 'WellBalance.SPPSIM']].rename(columns={'UtcTime':'time_key', 'WellBalance.SPPSIM':'spp'})
    model_spp_default['spp'] *= 1000

    return measure_ecd, model_ecd_default, model_ecd_correct, measure_spp, model_spp_default, model_spp_correct

def read_BDC_1():
    model_ecd_correct_path = r"C:\Users\Jyu20\Downloads\BDC_Run1_bha_WellBalance_20240909111924_mud_property.csv"
    model_ecd_correct = pd.read_csv(model_ecd_correct_path)[['time key', 'ecd at bit']]
    model_ecd_correct = model_ecd_correct.rename(columns={'time key':'time_key', 'ecd at bit':'ecd'})

    measure_ecd_path = r"C:\Users\Jyu20\Downloads\Channel_ECD.csv"
    measure_ecd = pd.read_csv(measure_ecd_path).rename(columns={'ecd_at_bit': 'ecd'})
    measure_ecd['ecd'] *= unit_convert('g/cm3', 'kg/m3')
    measure_ecd['time_key'] = pd.to_datetime(measure_ecd['time_key'])

    model_ecd_default_path = r"C:\Users\Jyu20\Downloads\BDC_Run1_bha_WellBalance_20240909143022_default_property.csv"
    model_ecd_default = pd.read_csv(model_ecd_default_path)
    model_ecd_default = model_ecd_default[['time key', 'ecd at bit']]
    model_ecd_default = model_ecd_default.rename(columns={ 'ecd at bit': 'ecd', 'time key': 'time_key'})

    return measure_ecd, model_ecd_default, model_ecd_correct

def read_BDC_402():
    rate = 100
    model_ecd_correct_path = r"C:\Users\Jyu20\Downloads\BDC-4-02_bha_run1_correct_mud_WellBalance_20240904160627_mud_property_correct_base_fluid.csv"
    model_ecd_correct = pd.read_csv(model_ecd_correct_path, skiprows=[1])[['time key', 'ecd at bit']]
    model_ecd_correct = model_ecd_correct.rename(columns={'time key':'time_key', 'ecd at bit':'ecd'})[::rate]

    model_spp_correct = pd.read_csv(model_ecd_correct_path, skiprows=[1])[['time key', 'simulated spp']]
    model_spp_correct = model_spp_correct.rename(columns={'time key':'time_key', 'simulated spp':'spp'})[::rate]

    measure_ecd_path = r"C:\Users\Jyu20\Downloads\Channels.csv"
    measure_ecd = pd.read_csv(measure_ecd_path, skiprows=[1]).rename(columns={'ECD_ARC_RT': 'ecd', 'Time': 'time_key'})[['time_key', 'ecd']].dropna()
    measure_ecd['time_key'] = pd.to_datetime(measure_ecd['time_key'])

    measure_spp = pd.read_csv(measure_ecd_path, skiprows=[1]).rename(columns={'STANDPIPE_PRS': 'spp', 'Time': 'time_key'})[['time_key', 'spp']].dropna()
    measure_spp['time_key'] = pd.to_datetime(measure_spp['time_key'])
    # print(measure_ecd)

    model_ecd_default_path = r"C:\Users\Jyu20\Downloads\BDC-4-02_bha_run1_correct_mud_WellBalance_20240903182817_master.csv"
    model_ecd_default = pd.read_csv(model_ecd_default_path, skiprows=[1])
    model_ecd_default = model_ecd_default[['time key', 'ecd at bit']]
    model_ecd_default = model_ecd_default.rename(columns={ 'ecd at bit': 'ecd', 'time key': 'time_key'})[::rate]
    model_ecd_default['time_key'] = pd.to_datetime(model_ecd_default['time_key'])

    model_spp_default = pd.read_csv(model_ecd_default_path, skiprows=[1])[['time key', 'simulated spp']].rename(columns={'simulated spp': 'spp', 'time key': 'time_key'})
    model_spp_default['time_key'] = pd.to_datetime(model_spp_default['time_key'])

    # measure_ecd = measure_ecd[measure_ecd['time_key'] > model_ecd_default['time_key'].iloc[0]]
    # measure_ecd = measure_ecd[measure_ecd['time_key'] < model_ecd_default['time_key'].iloc[-1]]


    return measure_ecd, model_ecd_default, model_ecd_correct, measure_spp, model_spp_default, model_spp_correct

def plot(measure_ecd, default_ecd, correct_ecd, well_name):
    marker_size = 4
    line_width = 1.5

    fig = go.Figure() 
    fig.add_trace(
        go.Scattergl(
            mode = 'lines+markers',
            marker = dict(
                size = marker_size
            ),
            line={'width': line_width},
            x=measure_ecd['time_key'], 
            y=measure_ecd['ecd'], 
            name="measure ecd",
            yaxis='y'),
    )

    fig.add_trace(
        go.Scattergl(
            mode = 'lines+markers',
            marker = dict(
                size = marker_size
            ),
            line={'width': line_width},
            x=default_ecd['time_key'], 
            y=default_ecd['ecd'], 
            name="model ecd with default mud property",
            yaxis='y'),
    )

    fig.add_trace(
        go.Scattergl(
            mode = 'lines+markers',
            marker = dict(
                size = marker_size
            ),
            line={'width': line_width},
            x=correct_ecd['time_key'], 
            y=correct_ecd['ecd'], 
            name="model ecd with correct mud property",
            yaxis='y'),
    )

    fig.update_xaxes(tickfont=dict(
                                    size=20,
                                    ) )

    fig.update_yaxes(tickfont=dict(
                                    size=20,
                                    ) )
    fig.update_layout(
        title = dict(
            text = fr'WellBalance Validation - {well_name}',
            font = dict(size = 30)
        ),
        yaxis=dict(title=dict(
            text = "ECD (kg/m3)",
            font = dict(size = 20)
            ), 
            )
            ,
        xaxis=dict(title=dict(
            text = "Time",
            font = dict(size = 20)
            )),
        legend = dict(font = dict(size = 20))
            )

    # fig.write_html(rf"C:\Users\Jyu20\Downloads\tnd_simulation_validation_{well_name}.html")
    fig.write_html('test_ecd.html')
    # fig.show()

def plot_spp(measure_spp, default_spp, correct_spp, well_name):
    marker_size = 4
    line_width = 1.5

    fig = go.Figure() 
    fig.add_trace(
        go.Scattergl(
            mode = 'lines+markers',
            marker = dict(
                size = marker_size
            ),
            line={'width': line_width},
            x=measure_spp['time_key'], 
            y=measure_spp['spp'], 
            name="measure spp",
            yaxis='y'),
    )

    fig.add_trace(
        go.Scattergl(
            mode = 'lines+markers',
            marker = dict(
                size = marker_size
            ),
            line={'width': line_width},
            x=default_spp['time_key'], 
            y=default_spp['spp'], 
            name="model spp with default mud property",
            yaxis='y'),
    )

    fig.add_trace(
        go.Scattergl(
            mode = 'lines+markers',
            marker = dict(
                size = marker_size
            ),
            line={'width': line_width},
            x=correct_spp['time_key'], 
            y=correct_spp['spp'], 
            name="model spp with correct mud property",
            yaxis='y'),
    )

    fig.update_xaxes(tickfont=dict(
                                    size=20,
                                    ) )

    fig.update_yaxes(tickfont=dict(
                                    size=20,
                                    ) )
    fig.update_layout(
        title = dict(
            text = fr'WellBalance Validation - {well_name}',
            font = dict(size = 30)
        ),
        yaxis=dict(title=dict(
            text = "SPP (Pa)",
            font = dict(size = 20)
            ), 
            )
            ,
        xaxis=dict(title=dict(
            text = "Time",
            font = dict(size = 20)
            )),
        legend = dict(font = dict(size = 20))
            )

    # fig.write_html(rf"C:\Users\Jyu20\Downloads\tnd_simulation_validation_{well_name}.html")
    fig.write_html('test_spp.html')
```

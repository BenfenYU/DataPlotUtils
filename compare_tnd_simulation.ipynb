```python

import pandas as pd
from utils import unit_convert
import plotly.graph_objects as go

well_name = '3s-617_Run3'
local_res_path = r"C:\Users\Jyu20\Downloads\3S-617_Run3_TnD.csv"
replay_res_path = r"C:\Users\Jyu20\Downloads\3s-617_run3_replayed.csv"

local_tnd_o = pd.read_csv(local_res_path)
local_tnd_o  = local_tnd_o.rename(columns={'TimeStamp': 'time_key', 'TND.HKLD_PU_MODEL': 'pick_up', 'TND.HKLD_SO_MODEL': 'slack_off'})[['time_key', 'pick_up', 'slack_off']].dropna()
local_tnd_o['time_key'] = pd.to_datetime(local_tnd_o['time_key']).apply(str)

replay_tnd_o = pd.read_csv(replay_res_path, skiprows=[1])
replay_tnd_o = replay_tnd_o.rename(columns={'UtcTime': 'time_key', 'TND.HKLD_PU_MODEL': 'pick_up', 'TND.HKLD_SO_MODEL': 'slack_off'})[['time_key', 'pick_up', 'slack_off']].dropna()
replay_tnd_o['time_key'] = pd.to_datetime(replay_tnd_o['time_key']).apply(str)
replay_tnd_o['pick_up'] *= unit_convert('kkgf', 'n')
replay_tnd_o['slack_off'] *= unit_convert('kkgf', 'n')
```

```python
sample_rate = 10


replay_tnd = replay_tnd_o.iloc[::sample_rate]
local_tnd = local_tnd_o.iloc[::sample_rate]

start_time = local_tnd['time_key'].iloc[0]
end_time = local_tnd['time_key'].iloc[-1]
replay_tnd = replay_tnd[replay_tnd['time_key'] > start_time]
replay_tnd = replay_tnd[replay_tnd['time_key'] < end_time]



'''
measured ecd at bit, simulated ecd at bit
'''

marker_size = 4
line_width = 0.5

fig = go.Figure() 
fig.add_trace(
    go.Scattergl(
        mode = 'markers',
        marker = dict(
            size = marker_size
        ),
        line={'width': line_width},
        x=replay_tnd['time_key'], 
        y=replay_tnd['pick_up'], 
        name="replayed pickup hkld",
        yaxis='y'),
)

fig.add_trace(
    go.Scattergl(
        mode = 'markers',
        marker = dict(
            size = marker_size
        ),
        line={'width': line_width},
        x=local_tnd['time_key'], 
        y=local_tnd['pick_up'], 
        name="local pickup hkld"),
)

fig.update_xaxes(tickfont=dict(
                                size=20,
                                ) )

fig.update_yaxes(tickfont=dict(
                                size=20,
                                ) )
fig.update_layout(
    title = dict(
        text = fr'Tnd Simulation Validation - {well_name}',
        font = dict(size = 30)
    ),
    yaxis=dict(title=dict(
        text = "Hookload (N)",
        font = dict(size = 20)
        ), 
        )
        ,
    xaxis=dict(title=dict(
        text = "Time",
        font = dict(size = 20)
        )),
    legend = dict(font = dict(size = 20))
        )

fig.write_html(rf"C:\Users\Jyu20\Downloads\tnd_simulation_validation_{well_name}.html")
fig.show()
```
